```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns=800

a=pd.read_csv('mitbih_train.csv', header=None)
a
```

```python
a.isnull()
```

```python
a.isnull().sum()
```

```python
a.value_counts()
```

```python
x=a.iloc[:,:-1]
x
```

```python
y=a.iloc[:,-1:]
y
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,shuffle=True,test_size=0.2,random_state=150)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
xtrain.shape
```

```python
model = Sequential()

model.add(Dense(units = 10, input_shape = (187,),activation = 'relu'))    # Input layer
          
model.add(Dense(units = 6, activation = 'relu'))    # hidden layer
          
model.add(Dense(1, activation = 'sigmoid')) # Output layer
          
          
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])  # compiling model
          
model.summary()
```

```python
model.fit(xtrain,ytrain,validation_split=0.1,epochs=20)
```

```python
model.predict(xtest)
```

```python
ytest
```

```python
model.evaluate(xtest,ytest)
```
